# Power BI DashBoard Requirements For the search term Report

Hereâ€™s a detailed explanation of each column in your **Sponsored Products Search Term Report**:

### 1. **Date**
   - **Definition**: The date when the search term data was recorded.
   - **Usage**: Useful for tracking performance over time, identifying seasonal trends, and understanding daily campaign impacts.

### 2. **Portfolio Name**
   - **Definition**: The name of the portfolio under which the campaign is organized.
   - **Usage**: Portfolios help organize campaigns, especially if there are multiple products or brands. This column can be used to analyze performance by portfolio.

### 3. **Currency**
   - **Definition**: The currency in which the costs, sales, and revenue metrics are measured.
   - **Usage**: Helps ensure consistency in financial reporting, especially if campaigns run across different regions with varying currencies.

### 4. **Campaign Name**
   - **Definition**: The name of the advertising campaign that contains the ad group and specific targeting details.
   - **Usage**: Useful for analyzing the performance of each campaign, comparing different campaigns, and identifying high-performing or underperforming campaigns.

### 5. **Ad Group Name**
   - **Definition**: The specific ad group within a campaign, containing ads targeted toward similar keywords or customer segments.
   - **Usage**: Allows for deeper analysis at the ad group level, which can be used to optimize specific ad groups within a larger campaign.

### 6. **Targeting**
   - **Definition**: Indicates whether the ad was shown based on keyword targeting (specific words or phrases) or product targeting (targeting specific products).
   - **Usage**: Provides insights into the effectiveness of different targeting methods and helps optimize ad targeting strategies.

### 7. **Match Type**
   - **Definition**: The type of keyword match used in the ad group: **Broad**, **Phrase**, or **Exact**.
     - **Broad**: Ads may appear for search terms containing the keyword in any order, plus related variants.
     - **Phrase**: Ads show for searches containing the exact phrase or close variations.
     - **Exact**: Ads show only for searches with the exact keyword, in the same order.
   - **Usage**: Useful for understanding which match type performs best, allowing advertisers to allocate budgets based on performance.

### 8. **Customer Search Term**
   - **Definition**: The specific search term entered by a customer that triggered the ad.
   - **Usage**: Provides valuable insight into customer search behavior and helps optimize keywords for better targeting.

### 9. **Impressions**
   - **Definition**: The number of times the ad was shown to users.
   - **Usage**: High impressions indicate high visibility, while low impressions may suggest the need for better targeting or higher bids.

### 10. **Clicks**
   - **Definition**: The number of times users clicked on the ad.
   - **Usage**: Indicates engagement level; high clicks typically suggest relevance to search terms and interest in the product.

### 11. **Click-Through Rate (CTR)**
   - **Definition**: The percentage of impressions that resulted in clicks.
     \[
     \text{CTR} = \frac{\text{Clicks}}{\text{Impressions}} \times 100
     \]
   - **Usage**: Higher CTR suggests that the ad is compelling and relevant, while a low CTR may indicate a need for better ad copy or targeting.

### 12. **Cost Per Click (CPC)**
   - **Definition**: The average amount spent per click on the ad.
     \[
     \text{CPC} = \frac{\text{Spend}}{\text{Clicks}}
     \]
   - **Usage**: Helps control budget by understanding the cost of each engagement. A high CPC might suggest competitive keywords.

### 13. **Spend**
   - **Definition**: The total amount spent on clicks for this search term or ad within the specified period.
   - **Usage**: Essential for budget management and assessing the cost-effectiveness of campaigns.

### 14. **7 Day Total Sales**
   - **Definition**: The total sales (in currency) generated within 7 days of customers clicking the ad.
   - **Usage**: Tracks how well clicks convert into revenue, allowing analysis of post-click sales behavior and campaign effectiveness.

### 15. **Total Advertising Cost of Sales (ACOS)**
   - **Definition**: The percentage of ad spend relative to the total sales generated.
     \[
     \text{ACOS} = \frac{\text{Spend}}{\text{7 Day Total Sales}} \times 100
     \]
   - **Usage**: A lower ACOS indicates a more cost-effective campaign. Useful for optimizing budget allocation.

### 16. **Total Return on Advertising Spend (ROAS)**
   - **Definition**: The total return generated for every dollar spent on advertising.
     \[
     \text{ROAS} = \frac{\text{7 Day Total Sales}}{\text{Spend}}
     \]
   - **Usage**: Higher ROAS suggests higher profitability, making it a key metric for assessing campaign success.

### 17. **7 Day Total Orders (#)**
   - **Definition**: The total number of orders placed within 7 days of clicking the ad.
   - **Usage**: Indicates purchase behavior post-click, providing insight into campaign effectiveness.

### 18. **7 Day Total Units (#)**
   - **Definition**: The total number of units sold within 7 days of the ad click.
   - **Usage**: Useful for understanding customer purchase volume after engaging with the ad.

### 19. **7 Day Conversion Rate**
   - **Definition**: The percentage of clicks that led to an order within 7 days.
     \[
     \text{Conversion Rate} = \frac{\text{7 Day Total Orders}}{\text{Clicks}} \times 100
     \]
   - **Usage**: Measures the effectiveness of clicks in driving conversions. A higher rate implies that the ad is leading to purchases.

### 20. **7 Day Advertised SKU Units (#)**
   - **Definition**: The number of units of the advertised SKU sold within 7 days after a click.
   - **Usage**: Helps measure sales for the specific product being advertised.

### 21. **7 Day Other SKU Units (#)**
   - **Definition**: The number of units of other SKUs (not the advertised SKU) sold within 7 days after a click.
   - **Usage**: Shows if the ad leads to purchases of related products, useful for cross-selling analysis.

### 22. **7 Day Advertised SKU Sales**
   - **Definition**: The revenue generated from the advertised SKU within 7 days of a click.
   - **Usage**: Focuses on the effectiveness of ads for the promoted product, showing the specific return on advertising for that SKU.

### 23. **7 Day Other SKU Sales**
   - **Definition**: The revenue generated from other SKUs (not the advertised SKU) within 7 days of a click.
   - **Usage**: Reveals if ads for one product drive interest in other products, useful for evaluating indirect sales impact.
